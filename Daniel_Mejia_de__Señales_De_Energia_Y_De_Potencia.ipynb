# Ejercicio 1

1. Graficar la potencia media de una impedancia $R+jwL$, cuándo es excitada por una fuente sinusoidal.

2. Graficar la potencia media de una impedancia $R-j\frac{1}{wC} $, cuándo es excitada por una fuente sinusoidal.

### 1. Circuito RL


````
import matplotlib.pyplot as plt
import numpy as np

w = 300
L = 0.01
R = 4
Imax = 2
T0 = 2*np.pi/w
N = 100 #número de muestras

t = np.linspace(0, T0, N)
i = Imax*np.sin(w*t)

Vmax = np.sqrt(np.square(R)+np.square(w*L))*Imax
theta = np.arctan(w*L/R)
v = Vmax*np.sin(w*t + theta)

plt.figure(figsize=[10,5])
plt.plot(t, i, label = "i(t)")
plt.plot(t, v, label = "v(t)")
plt.xlabel('t')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()

p = v*i #potencia instantanea
P = np.trapz(p, t)/T0 #Potencia media TAMBIEN SE PUEDE RESOLVER POR SYMPY
print("Potencia media: "+str(P))
````

````
plt.figure(figsize=[10,5])
plt.plot(t, p, label = "Potencia instantanea p(t)")
plt.plot(t, P*np.ones(len(t)), "g", label = "Potencia media P")
plt.xlabel('t')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
````

###2. Circuto RC

````

C = 0.01
R=4
w=300
Imax=2
T0 = 2*np.pi/w
N = 100 #número de muestras

t = np.linspace(0, T0, N)
i = Imax*np.sin(w*t)

Vmax = np.sqrt(np.square(R)+np.square(1/(w*C)))*Imax
theta = -np.arctan(1/(w*R*C))
v = Vmax*np.sin(w*t + theta)

plt.figure(figsize=[10,5])
plt.plot(t, i, label = "i(t)")
plt.plot(t, v, label = "v(t)")
plt.xlabel('t')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()

p = v*i #potencia instantanea
P = np.trapz(p, t)/T0 #Potencia media TAMBIEN SE PUEDE RESOLVER POR SYMPY
print("Potencia media: "+str(P))
````

## Uso del Valor RMS

El valor RMS de una señal es una medida del valor efectivo o equivalente de una corriente alterna o una tensión alterna. El RMS es especialmente útil en aplicaciones eléctricas y electrónicas para determinar la potencia de una señal alterna. La fórmula general para calcular el valor RMS de una función \( y(t) \) es:
\begin{equation}
\
\text{RMS} = \sqrt{\frac{1}{T} \int_0^T [y(t)]^2 \, dt}
\end{equation}


donde \( T \) es el periodo de la señal.

## Demostración del Valor RMS para una Onda Sinusoidal

Para una onda sinusoidal \
\begin{equation}
\
\ y = A_1 \sin(2 \pi f t) \:
\
\end{equation}
\begin{equation}
\
\text{RMS} = \sqrt{\frac{1}{T} \int_0^T [A_1 \sin(2 \pi f t)]^2 \, dt}
\
\end{equation}
\begin{equation}
\
\text{RMS} = \sqrt{\frac{A_1^2}{T} \int_0^T \sin^2(2 \pi f t) \, dt}
\
\end{equation}
Usando la identidad trigonométrica
\begin{equation}
\ \sin^2(x) = \frac{1 - \cos(2x)}{2} \:
\end{equation}
\begin{equation}
\
\text{RMS} = \sqrt{\frac{A_1^2}{T} \int_0^T \frac{1 - \cos(4 \pi f t)}{2} \, dt}
\
\end{equation}
\
\begin{equation}
\
\text{RMS} = \sqrt{\frac{A_1^2}{2T} \left[ T - \frac{\sin(4 \pi f T)}{4 \pi f} \right]}
\
\end{equation}

Dado que
\begin{equation}
\ \sin(4 \pi f T) = 0 \:
\end{equation}
\begin{equation}
\
\text{RMS} = \sqrt{\frac{A_1^2}{2T} \cdot T}
\
\end{equation}
\begin{equation}
\
\text{RMS} = \frac{A_1}{\sqrt{2}}
\end{equation}
Este valor es crucial para determinar la potencia efectiva de la señal en sistemas eléctricos y electrónicos.
