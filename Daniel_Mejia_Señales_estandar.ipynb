```python

%matplotlib inline
import sympy as sym
sym.init_printing()
```

```python
t, sigma, omega = sym.symbols('t sigma omega', real=True)#definir variables simbólicas
s = sigma + 1j*omega  #sobre carga operador para definir parte imaginaria
x = sym.exp(s*t)#función exponencial simbólica desde sympy
x

y = x.subs({omega: 3, sigma : -0.2}) #sustituir variables omega y sigma por valores numéricos
print(r'x(t)=',x)
#se gráfica desde el paquete plot de sympy por ser objetos de variable simbólica
sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',ylim=[-1,1])
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',ylim=[-1,1])
```

**Ejercicio**

* Pruebe otros valores de `omega` y `sigma` para crear señales con amplitud creciente/decreciente/constante y diferentes frecuencias angulares.  

```python
#creciente -> sigma > 0
y = x.subs({omega: 3, sigma : 0.7})
print(r'x(t)=',x)

sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',ylim=[-50,100]) #r' indica al linterprete una variable tipo string
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',ylim=[-60,100]);
```



Como se puede observar en las graficas, cuando sigma es mayor a cero, la función es creciente porque aumenta hasta irse al infinito. Esto significa que a medida que la variable independiente crece, el valor de la función también lo hace continuamente.

```python
#ejemplo constante sigma = 0
y = x.subs({omega: 3, sigma : 0})
print(r'x(t)=',x)

sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',ylim=[-1,1]) #r' indica al interprete una variable tipo string
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',ylim=[-1,1]);
```

Como se puede observar en las dos graficas anteriores, cuando sigma es cero, la función tiene una amplitud constante y comienza a oscilar sin variar la intensidad de esas oscilaciones. Este comportamiento es similar a la combinación de dos funciones seno y coseno, donde la amplitud de las oscilaciones se mantiene fija mientras que la función sigue un patrón periódico.

```python
#ejemplo decreciente sigma < 0
y = x.subs({omega: 3, sigma : -0.2})
print(r'x(t)=',x)

sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',ylim=[-1,1]) #r' indica al interprete una variable tipo string
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',ylim=[-1,1]);
```

Como se observa en las gráficas, para valores de sigma menores a cero, la función se vuelve decreciente y eventualmente alcanza un valor mínimo (aproximadamente cero).

**Ejercicio**

* Use la función $\text{rect}(t)$ para construir una función $\cos(\omega t)$ con $\omega=2$, la cual se enciende en $t=-\frac{1}{2}$ y se apaga en $t=+\frac{1}{2}$.

```python
class rect(sym.Function): #clase/objeto tipo sympy

    @classmethod #decorador objeto
    def eval(cls, arg):
        return sym.Heaviside(arg + sym.S.Half) - sym.Heaviside(arg - sym.S.Half)

x = sym.cos(omega*t) * rect(t)
sym.plot(x.subs(omega,2), (t, -2, 2), ylim=[-1.2, 1.2], ylabel=r'$x(t)$');

```

Genere un [rectificador de onda completa](https://en.wikipedia.org/wiki/Rectifier#Full-wave_rectification) utilizando la función $\text{sgn}(t)$
